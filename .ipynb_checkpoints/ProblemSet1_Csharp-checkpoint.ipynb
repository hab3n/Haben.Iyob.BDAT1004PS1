## Question 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. 
By Heron's formula, the area of a triangle with side lengths a, b, and c is   s(s-a)(s-b)(s-c), where   s=(a+b+c)/2.   >>> triangleArea(2,2,2) 1.7320508075688772 

```
using System;//Visual Studio Syntax
using System.Collections.Generic;
using System.Linq;
using System.Text;
{
    public class AreaOfTriangle
    {
        public static void Main()
        {
            //Get inputs of length of sides of triangle
            Console.Write("Enter the length of side a:");
            double a = Convert.ToDouble(Console.ReadLine());
           
            Console.Write("Enter the length of side b:");
            double b = Convert.ToDouble(Console.ReadLine());
            
            Console.Write("Enter the length of side c:");
            double c = Convert.ToDouble(Console.ReadLine());

            //s=(a+b+c)/2
            double s = (a + b + c) / 2;
            
            //Hanlan's formula: s(s-a)(s-b)(s-c)
            double Area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            Console.Write("Area of a Triangle = " + Area);
            Console.ReadKey();
        }
    }
}
```

## Question 4 C# 
Write a program in C# Sharp to separate odd and even integers in separate arrays. 
Go to the editor Test Data : 
Input the number of elements to be stored in the array :5 
Input 5 elements in the array : 
element - 0 : 25 
element - 1 : 47 
element - 2 : 42 
element - 3 : 56 
element - 4 : 32 
Expected Output: 
The Even elements are: 42 56 32 
The Odd elements are : 25 47 

```
using System;
using System.Security.Cryptography;
using System.Text;


public class OddsEven
{
    public static void Main()
    {

        //25,  47, 42, 56, 32;
        int[] arr = new int[5];
        int oddCount = 0;
        int evenCount = 0;
        Console.Write("Input 5 elements in the array:\n");
        for (int i = 0; i < arr.Length; i++)
        {
            Console.Write("Element[" + (i) + "]: ");
            arr[i] = Convert.ToInt32(Console.ReadLine());
            if (arr[i] % 2 != 0)
            {
                oddCount++;
            }
            else
            {
                evenCount++;
            }
        }
        //Store odd and even numbers into different array
        int[] oddNumbers = new int[oddCount];
        int[] evenNumbers = new int[evenCount];

        for (int i = 0; i < arr.Length; i++)
        {
            if (arr[i] % 2 != 0)
                oddNumbers[--oddCount] = arr[i];
            else
                evenNumbers[--evenCount] = arr[i];
        }

        Console.WriteLine("\nOdd Numbers:");
        for (int i = 0; i < oddNumbers.Length; i++)
        {
            Console.Write((oddNumbers[i] + " "));
        }

        Console.WriteLine("\nEven Numbers:");
        for (int i = 0; i < evenNumbers.Length; i++)
        {
            Console.Write(evenNumbers[i] + " ");
        }

    }
}
```

Question 5 C# 
a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False depending on whether the point (x,y) lies in the rectangle with lower left corner (x1,y1) and upper right corner (x2,y2).  

#>>> inside(1,1,0,0,2,3) 
True 
#>>> inside(-1,-1,0,0,2,3) 
False  

b. Use function inside() from part a. to write an expression that tests whether the point (1,1) lies in both of the following rectangles: one with lower left corner (0.3, 0.5) and upper right corner (1.1, 0.7) and the other with lower left corner (0.5, 0.2) and upper right corner (1.1, 2).  

```
using System;

class Inside
{

    // boolean function to see if the point lies within the rectangle
    static bool Check(double x1, double y1, double x2,
                        double y2, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)
            return true;

        return false;
    }
    
    public static void Main()
    {
        //get input for the point to test and coordinates for rectangle
        Console.Write("Enter value for x to test:");
        double x = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter the value for y to test:");
        double y = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter lower left value x1:");
        double x1 = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter lower left value y1:");
        double y1 = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter upper right value x2:");
        double x2 = Convert.ToDouble(Console.ReadLine());
        Console.Write("Enter upper right value y2:");
        double y2 = Convert.ToDouble(Console.ReadLine());


        // if statement calls 'point' function to check
        // if point x,y is within the rectagle or not
        if (Check(x1, y1, x2, y2, x, y))
            Console.Write("Yes, point " + (x, y) + " does lies in the rectangle.");
        else
            Console.Write("No");
    }
}
```

```
using System;

class Inside
{

    // boolean function to see if the point lies within the rectangle
    static bool Check(double x1, double y1, double x2,
                        double y2, double x3, double y3, 
                        double x4, double y4, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)

            if (x > x3 && x < x4 &&
            y > y3 && y < y4)
                return true;

        return false;
    }

    public static void Main()
    {
        //left corner (0.3, 0.5) and upper right corner (1.1, 0.7)
        //and the other with lower left corner (0.5, 0.2) and upper right corner (1.1, 2)
        double x1 = 0.3, y1 = 0.5,
        x2 = 1.1, y2 = 0.7, 
        x3 = 0.5, y3 = 0.2,
        x4 = 1.1, y4 = 2; ;

        // (1, 1)
        double x = 1, y = 1;

        // if statement calls 'point' function to check
        // if point x,y is within the rectagle or not
        if (Check(x1, y1, x2, y2, x3, y3, x4, y4, x, y))
            Console.Write("Yes, point " + (x, y) + " does lies in the rectangle.");
        else
            Console.Write("No");
    }
}
```
